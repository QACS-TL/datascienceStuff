```python
import matplotlib.pyplot as plt
data = [2,3,4,6,7,7,8,19]
```

```python
plt.figure(figsize=(8,6))
ax = plt.boxplot(data, vert=False)
plt.title("Boxplot for simple stats")
plt.xlabel("Years old")
plt.show()
```
