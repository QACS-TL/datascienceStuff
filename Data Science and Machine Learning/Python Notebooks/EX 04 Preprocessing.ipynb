```python
#identify duplicates, null values and outliers
#scaling, encoding, feature selection and engineering

# pandas.DataFrame.drop_duplicates
import pandas as pd
from scipy import stats
df = pd.read_csv("data/renfe_trains.csv")
```

```python
df.dropna(subset = ["vehicle_class", "price", "fare"], how="all", inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

OUTLIERS

```python
IQR = df["price"].quantile(0.75) - df["price"].quantile(0.25)

upper_bound = df["price"].quantile(0.75) + (1.5 * IQR)
lower_bound = df["price"].quantile(0.25) - (1.5 * IQR)

outliers = df[(df["price"] < lower_bound) | (df["price"] > upper_bound)]
print(outliers)

```

```python
(outliers["fare"].value_counts()/df["fare"].value_counts())
```

```python
(outliers["vehicle_class"].value_counts()/df["vehicle_class"].value_counts()) * 100
```

Training, Scaling and Validation
